```python
import pandas as pd

amazon = pd.read_csv('amazon_prime_titles.csv')
disney = pd.read_csv('disney_plus_titles.csv')
hulu = pd.read_csv('hulu_titles.csv')
netflix = pd.read_csv('netflix_titles.csv')

for i in range(len(hulu['rating'])):
    if 'min' in str(hulu['rating'][i]) or 'seasons' in str(hulu['rating'][i]) or 'season' in str(hulu['rating'][i])or 'Season' in str(hulu['rating'][i]):
        hulu['duration'][i] = hulu['rating'][i]
        hulu['rating'][i] = 'NaN'

for i in range(len(netflix['rating'])):
    if 'min' in str(netflix['rating'][i]) or 'seasons' in str(netflix['rating'][i]) or 'season' in str(netflix['rating'][i]):
        netflix['duration'][i] = netflix['rating'][i]
        netflix['rating'][i] = 'NaN'
```

```python
import matplotlib.pyplot as plt

fig1 = plt.figure(figsize=(10,10))
ax1 = fig1.add_subplot(2,2,1)
ax2 = fig1.add_subplot(2,2,2)
ax3 = fig1.add_subplot(2,2,3)
ax4 = fig1.add_subplot(2,2,4)

ax1.pie(amazon['type'].value_counts(),labels = amazon['type'].value_counts().index, autopct='%1.1f%%',colors=['r','g'])
ax1.set_title('amazon')
ax2.pie(disney['type'].value_counts(),labels = disney['type'].value_counts().index, autopct='%1.1f%%',colors=['r','g'])
ax2.set_title('disney')
ax3.pie(hulu['type'].value_counts(),labels = hulu['type'].value_counts().index, autopct='%1.1f%%',colors=['g','r'])
ax3.set_title('hulu')
ax4.pie(netflix['type'].value_counts(),labels = netflix['type'].value_counts().index, autopct='%1.1f%%',colors=['r','g'])
ax4.set_title('netflix')
fig1.suptitle('Types: Movie vs. TV Show')

fig1.show()

```

```python
from matplotlib.ticker import MaxNLocator

amazon['year_added'] = pd.to_datetime(amazon['date_added']).dt.year
disney['year_added'] = pd.to_datetime(disney['date_added']).dt.year
hulu['year_added'] = pd.to_datetime(hulu['date_added']).dt.year
netflix['year_added'] = pd.to_datetime(netflix['date_added']).dt.year

amazon_movie = amazon[amazon['type']=='Movie']
amazon_TV = amazon[amazon['type']=='TV Show']
disney_movie = disney[disney['type']=='Movie']
disney_TV = disney[disney['type']=='TV Show']
hulu_movie = hulu[hulu['type']=='Movie']
hulu_TV = hulu[hulu['type']=='TV Show']
netflix_movie = netflix[netflix['type']=='Movie']
netflix_TV = netflix[netflix['type']=='TV Show']

fig3 = plt.figure(figsize=(10,10))
ax1 = fig3.add_subplot(2,2,1)
ax2 = fig3.add_subplot(2,2,2)
ax3 = fig3.add_subplot(2,2,3)
ax4 = fig3.add_subplot(2,2,4)

#ax1.bar(amazon['year_added'].value_counts().index,amazon['year_added'].value_counts())
ax1.hist([amazon_movie['year_added'],amazon_TV['year_added']],label=['Movie','TV Show'])
ax1.legend()
ax1.set_title('amazon')
ax2.hist([disney_movie['year_added'],disney_TV['year_added']],label=['Movie','TV Show'])
ax2.legend()
ax2.set_title('disney')
ax3.hist([hulu_movie['year_added'],hulu_TV['year_added']],label=['Movie','TV Show'])
ax3.legend()
ax3.set_title('hulu')
ax4.hist([netflix_movie['year_added'],netflix_TV['year_added']],label=['Movie','TV Show'])
ax4.legend()
ax4.set_title('netflix')

plt.gca().xaxis.set_major_locator(MaxNLocator(integer=True))

fig3.suptitle('added contents per Year')
fig3.show()
```

```python
from wordcloud import WordCloud

def show_wordcloud(a):
    words = []

    for i in list(a['listed_in']):
        for j in i.split(", "):
            words.append(j)
    words = " ".join(words)
    pic = WordCloud(width = 400, height = 200, background_color="white",collocations=False)
    
    pic.generate(words)

    return pic

    # plt.imshow(pic, interpolation='bilinear')
    # plt.axis('off')
    # plt.title('')

# show_wordcloud(amazon)

fig_wc = plt.figure(figsize=(20,10))
ax1 = fig_wc.add_subplot(4,1,1)
ax2 = fig_wc.add_subplot(4,1,2)
ax3 = fig_wc.add_subplot(4,1,3)
ax4 = fig_wc.add_subplot(4,1,4)

pic = show_wordcloud(amazon)
ax1.imshow(pic)
ax1.set_title('amazon')
ax1.axis('off')
pic = show_wordcloud(disney)
ax2.imshow(pic)
ax2.set_title('disney')
ax2.axis('off')
pic = show_wordcloud(hulu)
ax3.imshow(pic)
ax3.set_title('hulu')
ax3.axis('off')
pic = show_wordcloud(netflix)
ax4.imshow(pic)
ax4.set_title('netflix')
ax4.axis('off')
```

```python
df_concat = pd.concat([amazon, disney, hulu, netflix])
df_concat.info()
```

```python
plt.barh(df_concat['country'].value_counts().index[:10],df_concat['country'].value_counts()[:10])
#plt.title('Top 10 countries all platform')
plt.show()
```

```python
countries = ['United States','India','United Kingdom','Japan','Canada','South Korea']

for i in countries:
    df_i = df_concat[df_concat['country'] == i]
    count = df_i.groupby(['release_year'])['show_id'].count()
    plt.plot(count.index, count, label = i)
plt.legend()
plt.title('Production per year')
```

```python
types = ['Movie', 'TV Show']

for i in types:
    df_i = df_concat[df_concat['type'] == i]
    count = df_i.groupby(['release_year'])['show_id'].count()
    plt.plot(count.index, count, label = i)
plt.legend()
plt.title('Production per year')
```
